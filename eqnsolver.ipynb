```
import tensorflow as tf
import numpy as np
import math
```

```
# y = (-x^2+(log(1+x^2)))

x = tf.Variable([399.0],trainable=True)
```

```
x.shape
```

```
def f(x):
  loss = abs(x**2+math.sin(x)-399) #minimizing -ve of the function (or maximizing the actual fn.)
  return loss

def gradients(x):
  with tf.GradientTape() as tape:
    z = f(x)
  gradient = tape.gradient(z,x)
  return gradient
```

```
optimizer = tf.keras.optimizers.Adam(learning_rate=0.1)#You can try other optimizers for more complex functions.
```

```
epochs = 1000#
```

```
for epoch in range(epochs):
  grad = gradients(x)
  optimizer.apply_gradients(zip([grad],[x]))
```
